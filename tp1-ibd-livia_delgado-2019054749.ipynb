 ### 1. Liste todos os códigos e nomes de funções da subfunção que possui o maior número de funções.

```python
query = "select cod_funcao, nome_funcao from funcao natural join detalhamento_funcao where cod_subfuncao in (select cod_subfuncao from (select cod_subfuncao, max(c) from (select cod_subfuncao, count(cod_funcao) as c from funcao natural join detalhamento_funcao group by cod_subfuncao)))"

df = pd.read_sql_query(query, conn)
df
```

### 5. Liste o código e nome dos órgãos que possuıram os 5 maiores valores de despesas totais nas funções de educação e saúde durante o ano de 2018.

```python
query = "select cod_orgao, nome_orgao, sum(valor_despesa) from despesa_publica natural join orgao natural join (select cod_detalhamento from detalhamento_funcao where cod_funcao in (select cod_funcao from funcao where nome_funcao IN ('Saúde', 'Educação'))) where ano_despesa = '2018' group by cod_orgao order by sum(valor_despesa) desc LIMIT 5"

df = pd.read_sql_query(query, conn)
df
```

### 6. Liste todos os nomes de funções e subfunções das despesas públicas dos ́ultimos 5 meses de 2018 e seus respectivos valores totais, ordenando de forma decrescente pelo valor.


```python
query = "select nome_funcao, nome_subfuncao, sum(valor_despesa) s from despesa_publica natural join detalhamento_funcao natural join funcao natural join subfuncao where mes_despesa > 7 and ano_despesa = 2018 group by nome_funcao, nome_subfuncao order by s desc"

df = pd.read_sql_query(query, conn)
df
```

### 7. Liste os programas de governo e a quantidade de  ́org ̃aos governamentais distintos envolvidos nas despesas públicas desses programas. Nesta consulta, desconsidere valores de despesa negativos ou iguais a zero no perıodo.


```python
query = "select distinct nome_programa_governo, count(cod_orgao) from programa_governo natural join (select * from despesa_publica where valor_despesa > 0) group by nome_programa_governo"

df = pd.read_sql_query(query, conn)
df
```

### 8. Liste os nomes das subfunções das despesas públicas envolvidas para cada órgão governamental, apenas nos casos em que o valor total das despesas excedeu 10 mil reais. Sua consulta deve retornar o nome do órgão, a subfunção da despesa, a quantidade de entradas de despesas públicas e o valor total dessas despesas.


```python
query = "select nome_orgao, nome_subfuncao, qtd_despesas, valor_total_despesas from orgao natural join (select count(cod_despesa) as qtd_despesas, sum(valor_despesa) as valor_total_despesas, cod_orgao, cod_detalhamento from despesa_publica group by cod_orgao having valor_total_despesas > 10000) natural join detalhamento_funcao natural join subfuncao"

df = pd.read_sql_query(query, conn)
df
```

### 10. Liste todas as despesas públicas (código da despesa, mês e ano da despesa, modalidade, valor), ordenadas de forma decrescente pelo valor, em que o programa de governo é aquele que possui o maior valor total em despesas durante o ano de 2018.


```python
query = "select cod_despesa, mes_despesa, ano_despesa, valor_despesa from despesa_publica where cod_programa_governo in (select cod_programa_governo from (select cod_programa_governo, max(s) from (select cod_programa_governo, sum(valor_despesa) as s from despesa_publica where ano_despesa = 2018 group by cod_programa_governo))) order by valor_despesa desc"

df = pd.read_sql_query(query, conn)
df
```
